```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('books.csv')
```

```python
df.shape
```

### Quantidade de livros por categoria

```python
fig, ax = plt.subplots(figsize=(15,5))
df.groupby(['category']).count()['name'].plot.bar(ax=ax)
plt.show()
```

### Medias por categoria

```python
df.groupby('category').mean()
```

### Quantidade de livros pelas notas

```python
fig, ax = plt.subplots(figsize=(15,5))
df.groupby(['stars']).count()['name'].plot.bar(ax=ax)
plt.show()
```

```python
df.groupby(['stars']).count()['name']
```

### Preço medio por categoria

```python
fig, ax = plt.subplots(figsize=(15,5))
df.groupby(['category']).mean()['price'].plot.bar(ax=ax)
plt.show()
```

### Quantidade de livros por categoria e nota

```python
fig, ax = plt.subplots(figsize=(15,5))
df.groupby(['category','stars']).count()['name'].unstack().plot(kind='bar', stacked=True, ax=ax)
plt.show()
```
